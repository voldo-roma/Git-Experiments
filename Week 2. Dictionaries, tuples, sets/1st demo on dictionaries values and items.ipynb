#### Q6. You are given a dictionary sales_data that represents the monthly sales data for a store:

Write a Python function to calculate and return the total sales for each month using the .values() method. The total sales for each month should be computed as the sum of all values (sales numbers) for that month.

```python
sales_data = {
    'January': [3500, 4200, 3800],
    'February': [4100, 3900, 3700],
    'March': [4800, 4300, 4100]
}
```

```python
#code retake
def monthly_sales():
  for v in sales_data.values():
    print(sum(v))
monthly_sales()
```

```python
#code
total_sales = 0
for k, v in sales_data.items():
  total_sales += sum(v)
  print(k,total_sales)
```

```python
#with values, not items
total_sales = 0
for v in sales_data.values():
  total_sales += sum(v)
  print(total_sales)
```

#### Q8. You have a dictionary inventory that stores information about items in a store's inventory:

Write a Python function that takes the inventory dictionary and calculates and returns the total value of the entire inventory (the sum of the values of each item) using the .items() method.

```python
inventory = {
    'item1': {'name': 'Widget', 'price': 10.99, 'quantity': 100},
    'item2': {'name': 'Gadget', 'price': 19.99, 'quantity': 50},
    'item3': {'name': 'Doodad', 'price': 5.99, 'quantity': 200}
}
print(inventory)
```

```python
#code with items
def total_cost():
    #I need to combine price and quantity from multiple dictionaries (outside the loop)
    total = 0
    for _, v in inventory.items():
      curr_cost = v['price'] * v['quantity']
      total += curr_cost
    print(total)
    print(curr_cost)
```

```python
#code with values
def total_cost():
    #I need to combine price and quantity from multiple dictionaries (outside the loop)
    total = 0
    for v in inventory.values():
      curr_cost = v['price'] * v['quantity']
      total += curr_cost
print(total)    #print(curr_cost)
```
