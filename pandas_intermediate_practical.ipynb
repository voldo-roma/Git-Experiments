#### Q1. Given the following DataFrame `student_data`, group the students by "Class" and calculate the maximum age in each class using `.groupby()` and `.agg()`. Display the result.

```python
import pandas as pd

data = {
    "Student": ["Alice", "Bob", "Charlie", "David", "Eve"],
    "Class": ["A", "B", "A", "C", "B"],
    "Age": [18, 19, 17, 20, 18]
}

student_data = pd.DataFrame(data)

grouped = student_data.groupby("Class")
display(grouped.agg("max"))
```
