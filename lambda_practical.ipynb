#### Q1. Write a `lambda` function that calculates the cube of a number. Then, use this `lambda` function with the `map()` function to calculate the cubes of numbers in a given list.

```python
# Lambda function to calculate the cube of a number

# Using the lambda function with map to calculate cubes
numbers = [1, 2, 3, 4, 5]
f = list(map(lambda x: x**3, numbers ))
print(f)
```

#### Q2. Given a list of words, filter this list to extract all the words that contain the letter "e." Use the filter() function with a lambda expression to achieve this.

```python
words = ["apple", "banana", "cherry", "date", "grape", "kiwi", "lemon"]
contains_e = lambda word: 'e' in word
e_words=list(filter(contains_e, words))
print(e_words)
```

#### Q3. Create a list of dictionaries, each representing a book with "title" and "price" keys. Use the `sorted()` function with a `lambda` function to sort the list of books in ascending order of price.

```python
books = [
    {"title": "Book A", "price": 25},
    {"title": "Book B", "price": 18},
    {"title": "Book C", "price": 30},
    {"title": "Book D", "price": 15}
]

# Sorting the list of books by price in ascending order
sorted_books = sorted(books, key=lamda book:book['price'], reverse=True)
print(sorted_books)
```

#### Q4. Create a `lambda` function that takes a number as input and returns "Positive" if the number is greater than zero, "Negative" if it's less than zero, and "Zero" if it's equal to zero. Apply this `lambda` function using the `map()` function to categorize a list of numbers.

```python
nums = [-2, 0, 5, -7, 10, 0]
categorize=lambda x: "Positive" if x > 0 else ("Negative" if x < 0 else "Zero")
applied = list(map(categorize, nums))
print(applied)
```

#### Q5. You have a list of strings, each containing a first name and a last name separated by a space. Use a `lambda` function with the `sorted()` function to sort the list of names alphabetically by last name. (Hint: You may need to split each string and access the last name.)

```python
names = ["Alice Smith", "Bob Johnson", "Charlie Brown", "David Davis"]
sorted_names = sort(names,keys=lambda name: name.split())
print(sorted_names)

test = "alice smith"
test_split = test.split(" "[])
print(test_split)
```
