### Q1. Create a DataFrame with three columns: 'Name', 'Age', and 'City', containing information about three of your friends.


```python
#code
import pandas as pd

data = { "Name": ["Jim", "Mary", "Lucy", "John", "Sam", "Will", "Will"],
         "Age": [45, 26, 15, 70, 53, 83, 72],
         "City": ["Los Angeles", "London", "Paris", "London", None, "London", "Paris"]
}

df = pd.DataFrame(data)
display(df)
```

### Q2. a) Display the first 5 rows of your DataFrame.
### b) How can you display the last 3 rows of your DataFrame?


```python
#code
display(df.tail(3))
```
