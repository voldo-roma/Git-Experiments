## What is pandas?
Pandas is a popular open-source Python library widely used for data manipulation and analysis. Since python is the computing language for DATA, pandas is one of the most popular libraries. It provides powerful data structures and data analysis tools that make it easier to work with structured data. It provides a convenient and efficient way to handle, manipulate, and analyze data in Python, making it an essential tool for any data-related tasks. Excel is limited compared to what python offers in flexibility and size

### Key features: 
1. **Series**: A Series is a one-dimensional labeled array in pandas. It is used to represent a single column of data in a DataFrame. Series can hold data of any data type, including numeric, string, and datetime.

2. **DataFrame**: The DataFrame is the primary data structure in pandas. It is a two-dimensional labeled data structure, similar to a spreadsheet or SQL table. DataFrames allow you to store and manipulate data with rows and columns, making it easy to perform various data operations.

3. **Performance**: Pandas is designed to handle large datasets efficiently. It optimizes operations and uses NumPy under the hood to ensure fast and memory-efficient computations.

4. **Data Cleaning**: Pandas offers a variety of functions to handle missing data, duplicate records, and outliers. It allows you to drop or fill missing values, remove duplicates, and perform data imputation efficiently.

5. **Data Transformation**: Pandas provides powerful data transformation capabilities, such as filtering, sorting, grouping, and aggregating data. These operations help in summarizing and analyzing data effectively.

6. **Data Input and Output**: Pandas supports reading and writing data from various file formats, including CSV, Excel, SQL databases, and more. This functionality facilitates the easy import and export of data to and from different sources. 

7. **Time Series Data**: Pandas offers specialized data structures and functions for handling time series data. It can handle time-based indexing, resampling, and window-based operations efficiently.

8. **Data Visualization**: Although pandas itself does not provide data visualization tools, it integrates seamlessly with popular visualization libraries like Matplotlib and Seaborn. This combination allows you to create insightful visualizations based on your data.

# Data Structures (Series & Dataframe)

The main components of pandas are the series and dataframe.

Series - a labeled one-dimensional array, acting as a basic component of Pandas data structures. It stores various data types, has an index for easy data access, and supports efficient operations like indexing and mathematical computations. Series are versatile, used independently or as DataFrame columns, and are crucial for data analysis and manipulation tasks in Python.

Dataframe - a two-dimensional data structure at the core of Pandas. ***It resembles a spreadsheet or SQL table, allowing data to be organized in rows and columnsIt organizes data like a table, with rows and columns.*** Each column can hold different data types, like numbers, strings, or dates. The DataFrame offers powerful tools for data manipulation, filtering, sorting, and analysis. It also seamlessly integrates with data visualization libraries, making it convenient to create informative plots. Thanks to its flexibility and user-friendliness, the Pandas DataFrame is a vital tool for data analysis and exploration in Python.

> TLDR: A Series is essentially a column, and a DataFrame is a multi-dimensional table made up of a collection of Series.

```python
# Creating a series 
data_dict = {'Name': ['John', 'Emma', 'David', 'Sophia', 'Michael', 'Carla']}
series_dict = pd.Series(data_dict)
print(type(series_dict))
display(series_dict)

```

```python
# Creating a dataframe
data = {'Name': ['John', 'Emma', 'David', 'Sophia', 'Michael', 'Carla'],
        'Age': [28, 32, 45, 36, 40, 39],
        'City': ['New York', 'Paris', 'London', 'Tokyo', 'Sydney', 'New York'],
        'Height': [176, 155, 190, 142, 183, 179],
        'Employed': [True, True, True, False, False, True]
}
df = pd.DataFrame(data)
df_copy = df.copy()
```

```python
# Displaying the DataFrame 
display(df)
```

```python
# Basic DataFrame operations
print("Basic DataFrame operations:")
print("Shape:", df.shape) # prints the df's rows by columns
print("Column names:", df.columns) # prints out the df's column names
print("Data types:", df.dtypes) # prints out the df's column datatypes
```

### Dtypes
Pandas dtypes specify the data types associated with the values in a DataFrame or Series. Each column in a DataFrame can have its own data type, which defines the kind of data that can be stored in that column.

1. Numeric data types:

    - int64: 64-bit integer, represents whole numbers.
    - float64: 64-bit floating-point number, represents decimal numbers.


2. Text data types:
    - object: Represents strings or mixed data types. When a column contains different types of data (e.g., strings and numbers), Pandas assigns the object dtype to that column.


3. DateTime data types:
    - datetime64: Represents date and time information with nanosecond precision.


4. Boolean data type:
    - bool: Represents Boolean values (True or False).


5. Categorical data type:
    - category: Represents a categorical variable with a fixed set of unique values. Categorical data is memory-efficient and allows for faster data processing.

Understanding and managing dtypes is crucial for effective data analysis, memory optimization, and performing operations on the data.

You can change the data types of columns using `astype()` method. 


```python
df['Height'] = df['Height'].astype('float64')

print(df.dtypes)
```

### Aggregating Data with Pandas

Aggregating data involves summarizing information within your DataFrame to gain insights. Pandas provides several aggregation functions that you can use to calculate statistics like the mean, minimum, maximum, and more.

#### Calculating Mean
The .mean() function calculates the average of a column or DataFrame. Let's find the average age and grade of our friends:

```python
# Calculate the average age
average_age = df['Age'].mean()

# Calculate the average grade
average_height = df['Height'].mean()

print("Average Age:", average_age)
print("Average Height:", average_height)
```

#### Finding Minimum and Maximum Values
You can use .min() and .max() to find the minimum and maximum values in a column. For example, let's find the youngest and oldest friend in our DataFrame:

```python
# Find the youngest friend
youngest_age = df['Age'].min()

# Find the oldest friend
oldest_age = df['Age'].max()

print("Youngest Age:", youngest_age)
print("Oldest Age:", oldest_age)
```

#### Counting Values
To count the number of occurrences of unique values in a column, use .value_counts(). Let's count how many friends live in each city:

```python
# Count the number of friends in each city
city_counts = df['City'].value_counts()

print("City Counts:")
print(city_counts)
```

# Adding and deleting columns

**Adding Columns:**

- To add a new column to a DataFrame, you can simply assign values to a new column name using the assignment operator (=).
- If the specified column name does not exist, Pandas will create a new column with that name and assign the values to it. If the column already exists, the values in the existing column will be replaced with the new values.

```python
df['Profession'] = ['Engineer', 'Artist', 'Doctor', 'Teacher', 'Lawyer', 'Manager'] 
# length of new columns must match existing dataframe
display(df)
new_df = df.pop('Profession')
display(new_df)
```

**Removing Columns:**

- To remove a column from a DataFrame, you can use the del keyword or the drop() method.
- Using the del keyword: You can delete a column using the del keyword followed by the DataFrame name and the column name within square brackets.
- Using the drop() method: The drop() method allows you to remove columns by specifying the column name and setting the axis parameter to 1. By default, the drop() method returns a new DataFrame without the specified column. If you want to modify the original DataFrame in place, you can set the inplace parameter to True.

```python
df['Profession'] = ['Engineer', 'Artist', 'Doctor', 'Teacher', 'Lawyer', 'Manager'] 
display(df)

df = df.drop(1, axis=0)
#del df['Profession']
display(df)
```

```python
# re-adding column
df['Profession'] = ['Engineer', 'Artist', 'Doctor', 'Teacher', 'Lawyer', 'Manager'] 
display(df)

df.drop('Profession', axis=1, inplace=True)
display(df)
```

### Filtering Data with Conditional Selection

You can use conditional expressions to filter data based on specific conditions. For example, you can filter rows where a particular column has values greater than a certain threshold or where a text column contains a specific substring.
> Conditional selection is often done using comparison operators (>, <, ==, !=, etc.) combined with logical operators (& for AND, | for OR, ~ for NOT).

Pandas will return a DataFrame containing only the rows that satisfy the given conditions.

```python
data = {
    'Name': ['John', 'Emma', 'David', 'Sophia', 'Michael', 'Carla'],
    'Age': [28, 32, 45, 36, 40, 39],
    'City': ['New York', 'Paris', 'London', 'Tokyo', 'Sydney', 'New York'],
    'Height': [176, 155, 190, 142, 183, 179],
    'Employed': [True, True, True, False, False, True]
}
# Notice how before the index has been ints (0-5), now with the index arg, the index is letters.
# 
df = pd.DataFrame(data, index=['A', 'B', 'C', 'D', 'E', 'F'])
```

```python
bool_mask = df['Age'] > 35
filtered_df = df[bool_mask]
print(filtered_df)
```

```python
display(df)
filtered_height_df = df[df['Height'] < 183]
display(filtered_height_df)
```

```python
# Select rows where 'Name' contains string 'o'
filtered_df_str = df[df['Name'].str.contains('o')]
print(filtered_df_str)
```

```python
# Combining Multiple Conditions:
filtered_combo_df = df[(df['Age'] > 35) & (df['City'] != 'London')]
print(filtered_combo_df)
```

# Sorting the DataFrame

In Pandas, the `.sort_values()` method is used to sort a DataFrame based on the values in one or more columns. It arranges the rows of the DataFrame in ascending or descending order, depending on the specified sorting criteria.

```python
sorted_df = df.sort_values('City', ascending=False)
display(sorted_df)
```

**Sorting by Multiple Columns:**
You can also sort a DataFrame based on multiple columns by passing a list of column names to the by parameter. The DataFrame will be first sorted by the first column in the list, and if there are ties, it will be sorted by the second column, and so on.

```python
sorted_df2 = df.sort_values(by=['City', 'Age'], ascending=[True, True])
display(sorted_df2)
```

# Handling missing values

We must distinguish between None, 0 and NaN here: 
- None has the data type 'NoneType' and is therefore a value, and we can use this as a placeholder before adding values
- 0 is an integer, and therefore a value, this shows that we have a response in the cell, but the response is 0
- NaN stands for 'Not a Number' and so denotes a MISSING value, although it has the type float

```python
data_with_nan = {
    'Name': ['John', 'Emma', 'David', None, 'Michael'],
    'Age': [28, 32, None, 36, 40],
    'City': ['New York', 'Paris', 'London', 'Tokyo', np.NaN]
}
```

```python
df_nan = pd.DataFrame(data_with_nan)
display(df_nan)
```

```python
print(df_nan.isnull().sum())
```

`.dropna()` method:

The .dropna() method is used to remove missing values (NaN) from a DataFrame. It drops rows or columns that contain at least one NaN value, depending on the axis parameter.

```python
df_dropped = df_nan.dropna(axis=0)
df_dropped
```

`.fillna()` method:

The .fillna() method is used to fill missing values (NaN) in a DataFrame with specified values or by using specific filling strategies.

```python
df_fill = df_nan.fillna({'Name': 'Unknown', 'Age': 0, 'City': 'Unknown'})
df_fill
```
