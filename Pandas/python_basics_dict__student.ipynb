# Python Dicts

#### In Python, a dictionary is a collection of key-value pairs. Each key is unique and maps to a specific value. Dictionaries are a fundamental data structure used for various purposes, such as storing configuration settings, representing real-world objects, and more. Dicts are defined using curly brackets  and items are separated by commas.

## Dict Creation

Creating a dictionary in Python involves specifying key-value pairs within curly braces {}. Each key is separated from its corresponding value by a colon :. Multiple key-value pairs are separated by commas ,. Here's the basic syntax for creating a dictionary:

```python
my_dict = {
    key1: value1,
    key2: value2,
    key3: value3,
    # ... (more key-value pairs)
}
```

```python
# define a dict
person = {
    'name': 'Alice',
    'age': 30,
    'city': 'New York'
}
print(person)
```

## Accessing Values
In Python dictionaries, you can access the values associated with keys using square brackets [] or the .get() method. Here's how to do it:

### Using Square Brackets [ ]:
You can access values in a dictionary by specifying the key within square brackets. Here's the syntax:
- my_dict: The dictionary from which you want to access the value.
- key: The key for which you want to retrieve the associated value.

value = my_dict[key]

#### Example:

```python
# Accessing values
print("Name:", person['name'])
print("Age:", person['age'])
print("City:", person['city'])
# print('not_a key:', person['height'])
```

### Using .get() Method:
The .get() method allows you to retrieve the value associated with a key in a dictionary without raising an error if the key is not found. Here's the syntax:
- my_dict: The dictionary from which you want to access the value.
- key: The key for which you want to retrieve the associated value.
- default_value (optional): The value to return if the key is not found. If not provided, it defaults to None

```python
value = my_dict.get(key, default_value)
```

#### Example

```python
# Accessing values using .get()
name = person.get('name')        
age = person.get('Age')    # None (key not found) as the correct key is 'age'
city = person.get('city', 'N/A')  
# height = person.get('height', 'N/A')
print(name)
print(age)
print(city)
# print(height)
```

## Modifying Values
Dictionaries are mutable, which means you can change their values. 

In Python dictionaries, you can modify the values associated with keys by simply referencing the key and assigning a new value. Here's the basic syntax for modifying values in dictionaries:
- my_dict: The dictionary in which you want to modify the value.
- key: The key for which you want to change the associated value.
- new_value: The new value you want to assign to the key.

```python
my_dict[key] = new_value
```

## Example

```python
# Modifying a value
person['age'] = 31
print("Updated Age:", person['age'])
# print(person)
```

## Adding and Removing Key-Value Pairs
You can add new key-value pairs to a dictionary and remove existing ones.

### Adding a Key-Value Pair:
To add a new key-value pair to an existing dictionary, you can use the following syntax:
- my_dict: The dictionary to which you want to add a new key-value pair.
- new_key: The key you want to add.
- new_value: The value associated with the new key.

```python
my_dict[new_key] = new_value
```

### Removing a Key-Value Pair:
To remove a key-value pair from a dictionary, you can use the del statement along with the key you want to remove. The syntax is as follows:
- my_dict: The dictionary from which you want to remove a key-value pair.
- key_to_remove: The key you want to remove.

#### Example

```python
person = {
    'name': 'Alice',
    'age': 30,
    'city': 'New York'
}
```

```python
# Adding a new key-value pair
person['job'] = 'Engineer'
print(person)

# Removing a key-value pair
del person['city']
print(person)
```

## Iterating Through a Dictionary
In Python dictionaries, the .keys() and .values() functions are used to extract information about the keys and values stored in a dictionary, respectively. Here's an overview of these functions:

### .keys() function: 
The .keys() function returns a view object that displays a list of all the keys in a dictionary. This view object is iterable, which means you can use it in loops or convert it to a list or another data structure if needed.

```python
person = {
    'name': 'Alice',
    'age': 30,
    'city': 'New York'
}

keys = person.keys()
print(keys)

for key in keys:
    print(key)

# for k in person.keys():
#     print(k)
```

### .values() Function:
The .values() function returns a view object that displays a list of all the values in a dictionary. Like .keys(), this view object is iterable.

```python
# Get a list of all values
values = person.values()
print(values)

for value in values:
    print(value)

# for v in person.values():
#     print(v)
```

#### Note
Both .keys() and .values() return view objects, not lists. 
To convert the view objects into lists, you can use the list() constructor. 

```python
print(keys)
print(type(keys))

my_list = list(keys)
print(my_list)
print(type(my_list))

# ls = []
# for k in person.keys():
#     ls.append(k)
# print(ls)
```

## The .items Function:
The .items() function returns a view object that contains a list of tuples, where each tuple represents a key-value pair from the dictionary. This view object is iterable and allows you to access both the keys and values in a convenient way.

```python
# Iterating through a dictionary
items = person.items()

for key, value in items:
    print(key + ":", value)
print(type(items))

# for k, v in person.items():
#     print(k + ":", v)
```

### Note:
The .items() view object returns tuples containing key-value pairs, allowing you to simultaneously access both keys and values during iteration.

To convert the view object into a list of tuples, you can use the list() constructor. For example: list(my_dict.items()).

The .items() function is particularly useful when you need to iterate through both keys and values in a dictionary or perform operations that involve both parts of the key-value pair.

## Combining Dictionaries with Counters
Using the dictionary methods mentioned above we can combine with additional functionality in loops.

In this example we will add a counter whilst we traverse the dictionary to workout a students highest grade & workout the average mark from the class



```python
student_grades = {
    'Alice' : [70, 80, 85],
    'John' : [40, 60, 65],
    'Joe' : [25, 30, 36]
}
```

```python
total_students = 0
highest_grade = {}

for k, v in student_grades.items():
    stu_max = max(v)
    highest_grade[k] = stu_max
    total_students += sum(v)

avg_mark = total_students / len(student_grades)
avg_mark = "{:.0f}".format(avg_mark)

print(f' The average mark of the class is : {avg_mark}')

print(highest_grade)



```
