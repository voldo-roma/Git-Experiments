# Encapsulation

Encapsulation in OOP promotes the idea of bundling data and behaviors related to a particular entity into a single unit known as a class. It allows us to hide the internal implementation details of an object and expose only those methods and properties that are necessary for its intended usage. This is achieved by providing access control mechanisms to protect the data from unwanted modifications and ensure that the object's state remains consistent.

##  Key Features of Encapsulation

**1. Data and Behavior Together**:  Encapsulation in Python means combining data and the functions that operate on that data within a single class. This keeps the related functionality together, making the codebase more organized and easier to manage.

**2. Access Control**:  Python's encapsulation allows us to specify access levels for class members: public, private, and protected.

> Public members can be accessed from outside the class, while private members are hidden and can only be accessed from within the class itself. This access control ensures that the internal implementation details are not exposed directly to the outside world, preventing unintended misuse of the class.

**3. Information Hiding**: Encapsulation enables information hiding, where the inner workings of an object are kept hidden from the outside world. This improves code structure, reduces dependencies, and allows for changes in internal implementation without affecting external code that uses the class.


## Privacy and Access Control with Private, Protected, and Public Methods

### Public Methods (Interfaces)

Public methods are accessible from outside the class and serve as the primary interface for interacting with an object. They are designed to be used by external code and are meant to provide essential functionality without exposing the internal implementation.

> Public methods are marked by their regular method names and do not use any special prefixes or suffixes.

For example:

```python
class Car:
    def __init__(self, make, model):
        self.make = make
        self.model = model

    def start_engine(self):
        # Code to start the car's engine
        return 'start engine'
        pass

    def accelerate(self, speed):
        # Code to accelerate the car to a given speed
        pass

my_car = Car('2015', 'tesla')
print(my_car.make)
my_car.start_engine()
```

In this example, `start_engine()` and `accelerate(speed)` are public methods that are accessible from outside the `Car` class.

### Protected Methods (Convention-based):

In Python, there is no strict access control for protected methods like in some other programming languages. However, there is a convention to use a single underscore prefix for methods and attributes that should be considered protected. This convention signals that these methods should not be accessed or modified by external code, although Python does not enforce this.

For example:

```python
class Animal:
    def __init__(self, name, species):
        self.name = name
        self._species = species  # Convention-based protected attribute

    def _get_taxonomy(self):
        # Code to obtain a list of zoological synonyms for the animal
        pass

    def _get_class(self):
        # Code to obtain the animal's class (mammals, birds, amphibians, reptiles, or fish)
        pass

cat = Animal('Maya', 'cat')
print(cat._species)
```

In this example, `_get_taxonomy()` and `_get_class()` are considered protected methods. Although external code can still access and call these methods, the single underscore serves as a gentle reminder that they are intended for internal use.

### Private Methods (Encapsulated)

Private methods are intended to be completely hidden from external code. In Python, to create a private method, you can use two underscores as a prefix for the method name. This naming convention tells Python to perform name mangling, effectively making the method name more complex to discourage external access.

Note: In Python, private members are "name-mangled" to make them harder to create subclasses that accidentally override private attributes and methods. The actual names get transformed, so accessing them directly from outside the class becomes more complex. (e.g. `_ClassName__privateMemberName`)

For example:

```python
class Student:
    def __init__(self, name, age):
        self.name = name
        self.__age = age

    def __get_age(self):
        # Private method to get the student's age
        return self.__age

    def get_info(self):
        # Public method to get the student's name and age
        return f"Name: {self.name}, Age: {self.__get_age()}"

student_1 = Student('Miruna', 78)
student_1.get_info()
```

In this example, `__get_age()` is a private method, and it can only be accessed within the `Student` class itself. Attempting to call `__get_age()` from outside the class will result in a name error.

### Summary

- Public methods are accessible from outside the class and serve as the primary interface for interacting with the object
- Protected methods (convention-based) use a single underscore prefix and are intended for internal use but can still be accessed externally (although it's not recommended)
- Private methods use a double underscore prefix and are intended to be hidden from external code entirely

## When to Use Private, Public, and Protected Methods and Attributes

### Public Methods and Attributes

Use Public Methods when you want to provide a clear and well-defined interface for external code to interact with your class. Public methods should represent the core functionality of the class and provide the essential operations that users of the class need to perform.

Use Public Attributes when the data needs to be directly accessible from outside the class without any additional processing. However, keep in mind that direct access to attributes can lead to potential issues in maintaining data integrity.

For example:

```python
class Circle:
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14 * self.radius ** 2
    
circle_instance = Circle(20)
circle_instance.area()
```

In this example, the `area()` method is a public method that provides a straightforward way for external code to calculate the area of the circle.

### Protected Methods and Attributes

Use Protected Methods and Attributes when you want to signal that certain methods or attributes are intended for internal use within the class or its subclasses. Although Python does not enforce strict access control for protected members, using a single underscore prefix is a convention to indicate their intended usage.

Protected methods and attributes can be useful for sharing common functionalities among subclasses without exposing them directly to external code.

For example:

```python
class Animal:
    def __init__(self, name, species):
        self.name = name
        self._species = species

    def _make_sound(self):
        # Protected method for making the animal's sound
        pass

    def _get_species(self):
        # Protected method to get the animal's species
        return self._species
    
class Dog(Animal):
    def __init__(self, name):
        # Call the parent class constructor with name and species for a dog
        super().__init__(name, "Canine")

    def _make_sound(self):
        return f'{self._species} make "Woof!"'


class Cat(Animal):
    def __init__(self, name):
        # Call the parent class constructor with name and species for a cat
        super().__init__(name, "Feline")

    def _make_sound(self):
        print(f'{self._species} make "Meow!"')
    
    def play_sound(self):
        self._make_sound()
 

# Create instances of Dog and Cat
dog = Dog("Rex")
cat = Cat("Whiskers")
cat.play_sound()

# # Access protected attributes and methods
# print(dog.name)          # Output: Rex

# print(dog._get_species())# Output: Canine

# print(cat.name)          # Output: Whiskers

# print(cat._get_species())# Output: Feline
```

In this example, `_make_sound()` and `_get_species()` are protected methods that are intended for internal use within the `Animal` class or its subclasses.


### Second Example

```python
class Person:
    def __init__(self, name, age):
        self._name = name  # protected attribute
        self._age = age    # protected attribute

    def _display_info(self):
        # protected method
        print(f"Name: {self._name}, Age: {self._age}")

class Student(Person):
    def __init__(self, name, age, student_id):
        super().__init__(name, age)
        self._student_id = student_id  # protected attribute

    def display_student_info(self):
        self._display_info()  # accessing protected method from superclass
        print(f"Student ID: {self._student_id}")

# Create a student object
student = Student("Alice", 20, "S1234")

# Access and display information
student.display_student_info()
```

### Private Methods and Attributes

Use Private Methods and Attributes when you want to completely hide certain functionalities and data from external code. Private members are intended to be used only within the class, ensuring strong encapsulation and preventing direct access from outside.

Private methods can help to break down complex operations into smaller, more manageable parts, without exposing those parts as separate public methods.

For example:

```python
class BankAccount:
    def __init__(self, account_number, balance):
        self._account_number = account_number  #protected attribute
        self.__balance = balance               #private attribute

    def __validate_transaction(self, amount):
        # Private method to validate if the transaction is allowed
        return amount > 0

    def deposit(self, amount):
        # Public method to deposit money into the account
        if self.__validate_transaction(amount):
            self.__balance += amount

    def withdraw(self, amount):
        # Public method to withdraw money from the account
        if self.__validate_transaction(amount) and self.__balance >= amount:
            self.__balance -= amount
        else:
            print("Insufficient funds or invalid amount")

    def get_balance(self):
        # Public method to check the account balance
        return self.__balance
    
my_bank_account = BankAccount("23471335923", 40)
# my_bank_account._BankAccount__validate_transaction(-90)   #BAD practice
my_bank_account.deposit(68)
my_bank_account.get_balance()
```

In this `example, __validate_transaction()` is a private method that is used internally by the `deposit()` method to ensure that the transaction is valid before updating the account balance.

```python
class Employee:
    __employee_count = 0  # Private class variable 
                          # intended to be accessed only within the class it is defined in.

    def __init__(self, name, salary):
        self.__name = name  # Private instance variable
        self.__salary = salary  # Private instance variable
        Employee.__employee_count += 1  # Accessing private class variable and increment it

    def __display_details(self):  # Private instance method
        print(f"Name: {self.__name}, Salary: {self.__salary}")

    def get_employee_count(self):  # Public method to access private class variable
        return Employee.__employee_count

    def display_employee_info(self):  # Public method
        self.__display_details()  # Accessing private instance method


# Creating employee objects
emp1 = Employee("John Doe", 50000)
emp2 = Employee("Miruna Nitu", 60000)

# Accessing private attributes and methods
# Attempting to access them directly from outside the class will raise an AttributeError

# Accessing public method to get employee count
print(f"Total Employees: {emp1.get_employee_count()}")

# Displaying employee information
emp1.display_employee_info()
emp2.display_employee_info()
```

### Example using public, private and protected members

```python
class Person:
    def __init__(self, name, age):
        self.name = name        # Public attribute
        self._age = age         # Protected attribute
        self.__email = None     # Private attribute

    def display_info(self):
        print(f"Name: {self.name}, Age: {self._age}, Email: {self.__email}")

    def set_email(self, email):     # setter 
        self.__email = email

class Student(Person):
    def __init__(self, name, age, student_id):
        super().__init__(name, age)
        self.student_id = student_id

    def display_info(self):
        super().display_info()
        print(f"Student ID: {self.student_id}")

    def update_email(self, email):
        self.set_email(email)

# Create a student object
student = Student("Alice", 20, "S1234")

# Access and display information
student.display_info()

# Try to access protected and private attributes directly (will raise AttributeError)
# print(student._age)        # Accessing protected attribute directly
# print(student.__email)     # Accessing private attribute directly

# Update email using public method
student.update_email("alice@example.com")

# Display updated information
student.display_info()
```

## Python Property Decorator – @property

In Python, the property decorator is a built-in function that provides a way to define methods that act like attributes. It allows you to customize how a class attribute is accessed, set, and deleted.

The property decorator is typically used to implement the concept of getters, setters, and deleters for class attributes, providing a controlled interface for accessing and modifying those attributes.

```python
class BankAccount:
    def __init__(self):
        self.balance = 0  # Private attribute

    @property                         #getter
    def balance(self):
        return self.__balance

    @balance.setter                   #setter - change current balance 
    def balance(self, amount):
        if amount >= 0:
            self.__balance = amount
        else:
            print("Invalid amount. Balance remains unchanged.")

    @balance.deleter                 
    def balance(self):
        print("Balance has been reset.")
        self.__balance = 0

# Creating a BankAccount object
account = BankAccount()
print(account.balance)

# Getting the initial balance
print(f"Initial Balance: {account.balance}")

# # Setting a new balance
account.balance = 1000

# # Getting the updated balance
print(f"Updated Balance: {account.balance}")

# # Attempting to set a negative balance
account.balance = -500

# # Getting the balance after the failed attempt
print(f"Balance after failed attempt: {account.balance}")

# # Resetting the balance
del account.balance

# # Trying to access balance after deletion
print(f"Balance after deletion: {account.balance}")
```

In this example, balance is a property that has a getter (@property), a setter (@balance.setter), and a deleter (@balance.deleter). This allows controlled access and modification of the __balance attribute.
